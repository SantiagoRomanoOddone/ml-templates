```python
from textblob import TextBlob
import matplotlib.pyplot as plt
import pandas as pd

# Sample social media posts
data = {
    'post': [
        "I love the new features of this product!",
        "This is the worst service I've ever experienced.",
        "I'm not sure how I feel about the latest update.",
        "Absolutely fantastic! Highly recommend it.",
        "Terrible experience, will never buy again.",
        "It's okay, not great but not bad either.",
        "I'm so happy with my purchase!",
        "Very disappointed with the quality.",
        "Neutral opinion, neither good nor bad.",
        "Best purchase I've made in a while!"
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to determine sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    # Classify polarity
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the function to the 'post' column
df['sentiment'] = df['post'].apply(get_sentiment)

# Count the occurrences of each sentiment
sentiment_counts = df['sentiment'].value_counts()

# Plot
sentiment_counts.plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('Sentiment Analysis of Social Media Posts')
plt.xlabel('Sentiment')
plt.ylabel('Number of Posts')
plt.xticks(rotation=0)
plt.show()
```
